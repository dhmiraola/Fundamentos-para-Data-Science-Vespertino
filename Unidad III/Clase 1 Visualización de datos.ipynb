### Scatter Plot 

```python
data.plot(kind="scatter", x="SemanasG",y="PesoRN")
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(x=data["SemanasG"],y=data["PesoRN"])
plt.xlabel("Semanas de gestación")
plt.ylabel("Peso del recién nacido")
plt.title("Relación entre las semanas de gestación y peso del recién nacido")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(x=data["PGDE"],y=data["PesoRN"])
plt.xlabel("Peso ganado después del embarazo")
plt.ylabel("Peso del recién ancido")
plt.title("Relación entre las semanas de gestación y peso del recién nacido")

plt.show
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(x=data["ConsTabaco"],y=data["Edadmadre"])
plt.xlabel("Consumo de tabaco")
plt.ylabel("Edad madre")
plt.title("Relación entre las semanas de gestación y peso del recién nacido")

plt.show
```
