```python
# Importamos la librería Pandas y la librería Matplotlib

import pandas as pd # Librería para análisis exploratorio
import matplotlib.pyplot as plt # Librería para la visualización de datos
import seaborn as sns # Librería para la visualización de datos
```

```python
data = pd.read_csv("customers.csv")
data.head(10)
```

```python
data.shape
```

## Gráfico de barra

```python
plt.subplots(figsize=(10,8))
plt.bar(x=data["Gender"],height = data["Balance"] )
plt.title("Balance según género")
plt.ylabel("Balance")
plt.xlabel("Género")
plt.show()
```

## Gráfico de Barra usando Seaborn

```python
plt.figure(figsize=(10,10))
sns.barplot(x = data["Geography"], y = data["Balance"], hue = data["Gender"])
plt.title("Balance por País y Género")
plt.ylabel("Balance")
plt.xlabel("País")
plt.show()
```

## Gráfico de calor o Heatmap
* Se usa para determinar la correlación entre las variables numéricas de la base de datos

```python
correl = data.corr() # Se crea la correlación antes de realizar el gráfico
sns.heatmap(correl, cmap= "bwr", vmin=-1, vmax=1) 
```

## Gráfico de líneas
* Se utiliza para mostrar tendencias y patrones

```python
mes = ["ene","feb","mar","abr","may","jun","jul"]
venta = [10,20,100,50,20,5,30]

plt.figure(figsize=(10,8))
plt.plot(mes,venta, color="red", linewidth = 3)
plt.title("Venta por mes")
plt.ylabel("Ventas")
plt.xlabel("Mes")

plt.show()
```
